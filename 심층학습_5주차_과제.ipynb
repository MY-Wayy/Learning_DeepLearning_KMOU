# Step 1 лІ мқҙмҠӨлқјмқё лӘЁлҚё мӢӨн–ү

```python
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
!git clone https://github.com/taehojo/data.git
```

```python
df = pd.read_csv('/content/data/sonar3.csv')

df
```

```python
X = df.iloc[:, 0:60]
y = df.iloc[:,60]

print(X.head())
print(y.head())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=True)
```

```python
# kкІ№ көҗм°Ё кІҖмҰқ
from sklearn.model_selection import KFold
```

```python
k = 5
kfold = KFold(n_splits=k, shuffle=True)
acc_score = []
```

```python
def model_fn():
  model = Sequential()
  model.add(Dense(24, input_dim=60, activation='relu'))
  model.add(Dense(10, activation='relu'))
  model.add(Dense(1, activation='sigmoid'))
  model.summary()
  return model
```

```python
for train_idx, test_idx in kfold.split(X):
  X_train, X_test = X.iloc[train_idx, : ], X.iloc[test_idx, :]
  y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

  model = model_fn()
  model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
  model.fit(X_train, y_train, epochs=200, batch_size=10)

  accuracy = model.evaluate(X_test, y_test)[1]
  acc_score.append(accuracy)
```

```python
print('acc_score:',acc_score)
print('нҸүк·  м •нҷ•лҸ„:',sum(acc_score) / k)
```

# Step 3 - кІ°кіј л№„көҗн‘ң мһ‘м„ұ


```python
# н‘ң м¶ңл Ҙ мҪ”л“ң
import pandas as pd
import numpy as np

# Step 1 л°Ҹ Step 2м—җм„ң мғқм„ұлҗң к°Ғ лӘЁлҚёмқҳ accuracy лҰ¬мҠӨнҠё лӘЁмқҢ
acc_scores = [acc_score, acc_score_1, acc_score_2, acc_score_3, acc_score_4, acc_score_5]

# к°Ғ model_fn м •мқҳ л¶Җл¶„м—җм„ң м¶”м¶ңн•ң лӘЁлҚё кө¬мЎ° ('nn-nn-nn...' нҳ•мӢқ)
structures = [
    '24-10-1',                      # model_fn (Model 0)
    '64-32-1',                      # model_fn1
    '24-18-10-1',                   # model_fn2
    '24-10-1',                      # model_fn3
    '64-36-18-6-1',                 # model_fn4
    '24-18-10-8-1'                  # model_fn5
]

# к°Ғ model_fn м •мқҳ л¶Җл¶„м—җм„ң м¶”м¶ңн•ң нҷңм„ұнҷ” н•ЁмҲҳ
activations = [
    'relu-relu-sigmoid',            # model_fn (Model 0)
    'relu-relu-sigmoid',            # model_fn1
    'relu-relu-relu-sigmoid',       # model_fn2
    'tanh-tanh-sigmoid',            # model_fn3
    'relu-relu-relu-relu-sigmoid',  # model_fn4
    'tanh-tanh-tanh-tanh-sigmoid'   # model_fn5
]

# н‘ң мһ‘м„ұмқ„ мң„н•ң л”•м…”л„ҲлҰ¬ лҚ°мқҙн„° кө¬м„ұ
data = {
    'лӘЁлҚёлІҲнҳё': [0, 1, 2, 3, 4, 5],
    'лӘЁлҚё кө¬мЎ°': structures,
    'нҷңм„ұнҷ” н•ЁмҲҳ': activations,
    'fold1': [score[0] for score in acc_scores],
    'fold2': [score[1] for score in acc_scores],
    'fold3': [score[2] for score in acc_scores],
    'fold4': [score[3] for score in acc_scores],
    'fold5': [score[4] for score in acc_scores],
    'нҸүк·  м •нҷ•лҸ„': [sum(score) / k for score in acc_scores],  # мҡ”мІӯн•ҳмӢ  кі„мӮ°мӢқ м Ғмҡ©
    'н‘ңмӨҖнҺём°Ё': [np.std(score) for score in acc_scores]      # numpyмқҳ н‘ңмӨҖнҺём°Ё н•ЁмҲҳ мӮ¬мҡ©
}

# лҚ°мқҙн„°н”„л Ҳмһ„ мғқм„ұ л°Ҹ м¶ңл Ҙ
df_results = pd.DataFrame(data)
df_results
```

# Step 4 - мөңм Ғ лӘЁлҚё м„ м • л°Ҹ к·јкұ° м„ңмҲ 


### нҸүк·  м •нҷ•лҸ„ мёЎл©ҙ : **лӘЁлҚё 1**
лӘЁлҚё 1мқҖ мёө мҲҳлҠ” к·ёлҢҖлЎң н•ҳкі  к°Ғ мёөмқҳ л…ёл“ң мҲҳлҘј лҠҳлҰ° лӘЁлҚёмқҙлӢӨ. к°Ғ нҸҙл“ңмқҳ м •нҷ•лҸ„ к°’мқҙ л№„көҗм Ғ лҶ’кІҢ лӮҳмҷҖ нҸүк· м№ҳлҸ„ лҶ’кІҢ лӮҳмҷ”мңјлҜҖлЎң, нҸүк·  м •нҷ•лҸ„ лҳҗн•ң лҶ’кІҢ лӮҳмҷ”кё° л•Ңл¬ём—җ мқҙ лӘЁлҚёлЎң м„ м •н•ҳмҳҖлӢӨ.

### н‘ңмӨҖнҺём°Ё(м•Ҳм •м„ұ) мёЎл©ҙ : **лӘЁлҚё 1**
н‘ңмӨҖнҺём°Ё (м•Ҳм •м„ұ) лҳҗн•ң лӘЁлҚё 1 мқҙ м ңмқј мӨҖмҲҳн•ҳмҳҖлӢӨ. нҸүк· кіј к°Ғ нҸҙл“ңмқҳ м •нҷ•лҸ„ к°’ м°Ёмқҙк°Җ л№„көҗм Ғ нҒ¬кІҢ лӮҳм§Җ м•Ҡм•ҳлӢӨлҠ” кІғмқ„ м•Ң мҲҳ мһҲлӢӨ. мқҙлҠ” н•ҙлӢ№ лӘЁлҚём—җ н•ҷмҠөн•ң лҚ°мқҙн„°мҷҖ нҸүк°Җ лҚ°мқҙн„°к°Җ лӢӨлҘҙлҚ”лқјлҸ„, н•ҙлӢ№ лҚ°мқҙн„°лЎң мҳӨлІ„н”јнҢ…мқ„ л№„көҗм Ғ лҚң н•ңлӢӨлҠ” м§Җн‘ңлЎң ліј мҲҳ мһҲм—ҲлӢӨ.

### нҸүк·  м •нҷ•лҸ„мҷҖ н‘ңмӨҖнҺём°Ёк°Җ мғҒм¶©н•ҳлҠ” лӘЁлҚё : **лӘЁлҚё 3**
лӘЁлҚё 3лҠ” л…ёл“ң мҲҳ, мёө мҲҳлҠ” к·ёлҢҖлЎң н•ҳкі  м¶ңл Ҙмёөмқ„ м ңмҷён•ң мёөмқҳ нҷңм„ұнҷ” н•ЁмҲҳлҘј `tanh` лЎң л°”кҫј лӘЁлҚёмқҙлӢӨ. мқҙ лӘЁлҚёмқҖ м •нҷ•лҸ„к°Җ м „мІҙ лӘЁлҚё мӨ‘ 2мң„ мқҙм§Җл§Ң, н‘ңмӨҖнҺём°ЁлҠ” мөңн•ҳмң„мқҙлӢӨ.
м „мІҙм ҒмңјлЎң нҸҙл“ң лӢ№ м •нҷ•лҸ„к°Җ 80мқҙмғҒмқҙкұ°лӮҳ 90мқҙмғҒмқҙм§Җл§Ң, 5лІҲм§ё нҸҙл“ңм—җм„ң 68%мқҳ м •нҷ•лҸ„лҘј ліҙм—¬мЈјм—ҲлӢӨ. мқҙлҠ” 5лІҲм§ё нҸҙл“ңмқҳ лҚ°мқҙн„°к°Җ н•ҷмҠөм—җ л§Өмҡ° л¶ҲлҰ¬н•ң лҚ°мқҙн„° мҳҖмқҢмқ„ м•Ң мҲҳ мһҲлӢӨ. нҠ№м • лҚ°мқҙн„°лЎңлҠ” лҶ’мқҖ м •нҷ•лҸ„лҘј ліҙм—¬мЈјм§Җл§Ң, нҠ№м • лҚ°мқҙн„°лЎңлҠ” л§Өмҡ° лӮ®мқҖ м •нҷ•лҸ„лҘј ліҙм—¬мЈјлҠ” л¶Ҳм•Ҳм •н•ң лӘЁлҚёмқҙлқјкі  нҢҗлӢЁлҗҳм—ҲлӢӨ.
