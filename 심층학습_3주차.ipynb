# н…җм„ңн”ҢлЎңм—җм„ң мӢӨн–үн•ҳлҠ” лӢӨмӨ‘ нҡҢк·Җ мӢӨмҠө

```python
import numpy as np
import matplotlib.pyplot as plt

# н…җм„ңн”ҢлЎңмҡ°мқҳ мјҖлқјмҠӨ API
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
x = np.array([[2, 0], [4, 4], [6, 2], [8, 3]])
y = np.array([81, 93, 91, 97])
```

```python
#лҚ°мқҙн„°к°Җ л§Һмқ„ кІҪмҡ°, column row мҲҳк°Җ мңЎм•ҲмңјлЎң к°ҷмқҖм§Җ л№„көҗк°Җ нһҳл“ңлҜҖлЎң н•ЁмҲҳ мӮ¬мҡ©
x.shape, y.shape
```

```python
model = Sequential()

model.add(Dense(1, input_dim=2, activation="linear"))
model.summary()
```

```python
model.compile(loss="mse", optimizer="sgd")
```

```python
model.fit(x, y, epochs=500)
```

```python
#н•ҷмҠөлҗң лӘЁлҚём—җ к°’мқ„ л„Јм–ҙм„ң мҳҲмёЎ
import tensorflow as tf
input_data = tf.constant([[6,3]])
model.predict(input_data)
```

# н…җм„ңн”ҢлЎңм—җм„ң мӢӨн–үн•ҳлҠ” лЎңм§ҖмҠӨнӢұ нҡҢк·Җ (н•©кІ©/л¶Ҳн•©кІ©)

```python
x1 = np.array([2, 4, 6, 8, 10, 12, 14])
y1 = np.array([0, 0, 0, 1, 1, 1, 1])
```

```python
x1.shape, y1.shape
```

```python
model = Sequential()

model.add(Dense(1, input_dim=1, activation="sigmoid"))
model.summary()
```

```python
model.compile(loss="binary_crossentropy", optimizer="sgd")
```

```python
model.fit(x1, y1, epochs=500)
```

```python
pred = model.predict(x1)
plt.scatter(x1, y1)
plt.plot(x1, pred, 'r')
pred
```
